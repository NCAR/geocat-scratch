```python
import numpy as np
import xarray as xr

import xugrid as xu
```

```python
nodes = np.array([[0,1], [1,0], [2,2], [3,0], [4,1]])
faces = np.array([[0,1,2], [1,2,3], [2,3,4]])
fill_value = -1
```

```python
grid = xu.Ugrid2d(nodes[:, 0], nodes[:, 1], fill_value, faces)
da = xr.DataArray(
    data=[1.0, 2.0, 3.0],
    dims=[grid.face_dimension],
)
uda = xu.UgridDataArray(da, grid)
uda
```

```python
uda.ugrid.plot.line()
```

```python
uda.ugrid.to_netcdf("3faces.nc")
```

```python
uda.ugridto_netcdf("3faces.nc")
```
